### Data Preprocessing

Creating new columns and standardization the data

```python
df2['un_in_std'] = data['u_in'].transform(lambda x: (x - x.mean()) / x.std())
df2['time_step_std'] = data['time_step'].transform(lambda x: (x - x.mean()) / x.std())
```

```python
df2['time_step_after'] = data[['time_step_std']].shift()
df2['u_in_after'] = data[['un_in_std']].shift()
df2['u_out_after'] = data[['u_out']].shift()
df2['time_step_back'] = data[['time_step_std']].shift(-1)
df2['u_in_back'] = data[['un_in_std']].shift(-1)
df2['u_out_back'] = data[['u_out']].shift(-1)
df2['time_step_after2'] = data[['time_step_std']].shift(2)
df2['u_in_after2'] = data[['un_in_std']].shift(2)
df2['u_out_after2'] = data[['u_out']].shift(2)
df2['time_step_back2'] = data[['time_step_std']].shift(-2)
df2['u_in_back2'] = data[['un_in_std']].shift(-2)
df2['u_out_back2'] = data[['u_out']].shift(-2)
df2['time_step_after3'] = data[['time_step_std']].shift(3)
df2['u_in_after3'] = data[['un_in_std']].shift(3)
df2['u_out_after3'] = data[['u_out']].shift(3)
df2['time_step_back3'] = data[['time_step_std']].shift(-3)
df2['u_in_back3'] = data[['un_in_std']].shift(-3)
df2['u_out_back3'] = data[['u_out']].shift(-3)
df2['time_step_after4'] = data[['time_step_std']].shift(4)
df2['u_in_after4'] = data[['un_in_std']].shift(4)
df2['u_out_after4'] = data[['u_out']].shift(4)
df2['time_step_back4'] = data[['time_step_std']].shift(-4)
df2['u_in_back4'] = data[['un_in_std']].shift(-4)
df2['u_out_back4'] = data[['u_out']].shift(-4)
df2['time_step_after5'] = data[['time_step_std']].shift(5)
df2['u_in_after5'] = data[['un_in_std']].shift(5)
df2['u_out_after5'] = data[['u_out']].shift(5)
df2['time_step_back5'] = data[['time_step_std']].shift(-5)
df2['u_in_back5'] = data[['un_in_std']].shift(-5)
df2['u_out_back5'] = data[['u_out']].shift(-5)
```

```python
features = df2.drop(['pressure', 'id', 'breath_id', 'u_in', 'time_step'], axis = 1)
target = df2['pressure']
```

```python
features = np.array(features)
features.shape
```

```python
target = np.array(target)
target.shape
```

```python
target = target.reshape(13981, 75)
target.shape
```

```python
import tensorflow as tf
import random
```

```python
norm = tf.keras.layers.Normalization(input_shape = [75, features.shape[2],], axis = -1)
norm.adapt(features)
```

### DIAGRAMS FOR DEEP LEARNING ALGORITHMS

This one is RNN Deep Learning Algorithm Diagram

This one is LSTM Deep Learning Algorithm Diagram

IN LSTM (LONG SHORT TERM MEMORY): In this first we provide input i.e(xt) then will apply some activation function on it 
and then it will approch to (ct cell) then ct cell will pass the information to forget gate (basically forget gate will decide which information to keep and which not)
and then forget gate will provide the relvant/ important information to output gate 
and then output gate will provide the output i.e(ht)


This one is bi-directional LSTM Deep Learning Algorithm Diagram

```python
m = tf.keras.Sequential([
    norm,
    tf.keras.layers.Conv1D(128, 3, activation = "relu"),
    tf.keras.layers.MaxPooling1D(),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv1D(256, 3, activation = "relu"),
    tf.keras.layers.MaxPooling1D(),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128, return_sequences = True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128, return_sequences = True)),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(75,)
])
m.compile(optimizer=tf.optimizers.Adam(learning_rate=0.001), loss= "mae")
```

1. Convulation technique is used for feature extraction and also speically is used for image, text processing. 
2. 128, 3 means we have used 128 filters of size 3 each.
3. relu stands for rectified linear unit ( if we provide the input in positive array, it will give output as positive array and if we provide the input as negative array, it will give output as 0)
4. Max Pooling is used to extract maximum feature from my data 

```python
m.summary()
```

```python
history = m.fit(features, target, validation_split = 0.2, epochs = 100, batch_size = 512,
                callbacks = [tf.keras.callbacks.EarlyStopping(patience = 26, 
                                                              monitor = 'val_loss', 
                                                              mode = 'min', 
                                                              restore_best_weights=True)])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (20, 5))
plt.plot(history.history['loss'], label = "loss")
plt.plot(history.history['val_loss'], label = "val_loss")
plt.legend()
plt.show()
```

```python
m.evaluate(features, target)
```

```python
def draw_result(start, end, target, y_pred):
    plt.figure(figsize = (20, 7))
    plt.plot(np.reshape(target[start:end], -1), linewidth=5, label = "actual Pressure values")
    plt.plot(np.reshape(y_pred[start:end], -1), linewidth=2, label = "predict Pressure values")
    plt.legend()
```

```python
def create_table(target, y_pred):
    target = np.reshape(target, -1)
    y_pred = np.reshape(y_pred, -1)
    return pd.DataFrame({"actual Pressure values": target,
                         "predict Pressure values": y_pred,
                         "diff": np.abs(target - y_pred)})
```
